**CS1**. Create a class called `Square`. Its initialization takes in a variable which is the side of a square. The object has one attribute which is called `side` to represent the side of a square. It has one property called `area` which sets the dimension of the square. Setting the area changes the side of a square.

```python
# SQUARE
# attributes:
# _side
# properties:
# area, side
# methods:
# __str__

class Square_alternative:
    def __init__(self, length_of_side: Number=1) -> None:
        self._side = length_of_side 
    def side(self) -> Number:
        return self._side
    def area(self) -> Number:
        return (self._side)**2 
    def set_area(self, value: Number) -> None:
        self._area = value
        self._side = value ** 0.5
    def __str__(self) -> str:
        return f'Square with side {self.side}'

square_sample = Square_alternative(5)
print(square_sample.area()) # since area is no longer a property, i need to call it like a regular function
## to set area, i also need to call it like a regular function
square_sample.set_area(36)

class Square:
    def __init__(self, length_of_side: Number=1) -> None:
        self._side = length_of_side 
    @property
    def side(self) -> Number:
        return self._side
    @property
    def area(self) -> Number:
        return (self._side)**2 
    @area.setter
    def area(self, value: Number) -> None:
        # usually a setter cleans up the input and sets the instance attributes accordingly
        self._area = value
        self._side = value ** 0.5
    def __str__(self) -> str:
        return f'Square with side {self.side}'

square_1 = Square(5)
square_1.area = 36 # this will call the area setter func under @area.setter, and pass square_1 as self, plus 36 as the second arg (value)
print(square_1.side) 
print(square_1.area) # calling the area getter
```

```python
s: Square = Square()
assert s.area == 1

s: Square = Square(2)
assert s.area == 4
s.area = 9
assert s.side == 3
```

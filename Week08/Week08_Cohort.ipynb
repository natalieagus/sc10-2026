**CS1.** Define an Abstract Class for a State Machine, called `StateMachine`. The class has one attribute and one computed property:
- `state`: which is the current state of the machine and is the attribute of any `StateMachine` object instance.
- `start_state`: which is the initial state of the machine and is the computed property to be defined in the child class.

The class should define the following methods:
- `start()`: this method set the `state` property using the value in `start_state`. Once `state` has a value, the machine is considered started.
- `step(inp)`: this method takes in the current input and returns the current output. This method should move the state machine to the next state based on the current input and its current state. You should call `get_next_values(state, inp)` in your implementation.
- `done(state)`: this method always return `False`. A child class can override thid method to give a different condition to end the state machine.
- `is_done()`: is to be used internally to check if the state machine should terminate or not. This method simply calls `done(state)` and pass on the current `state`. The method `transduce(inp_list)` calls this method to check if it should terminates or not.
- `transduce(inp_list)`: this method calls `start()` to initialize the `state` with the `start_state` and run the state machine by calling `step(inp)` for every item in the `inp_list`. The method runs the state machine and produces the output list according to the number of input in the `inp_list` or when the state machine terminates according to the output of `is_done()` method. This method should call `is_done()` to see if it should terminate at a particular state.

This class should be an Abstract Class. Implement the following way:
- `SM` class inherits from `abc.ABC`, which is Python's Abstract Base Class (ABC). 
- Any implementation of State Machine's instances must declare the following property: `start_state`.
- Any implementation of State Machine's instances must implement the following abstract method: `get_next_values()` that takes in the current `state` and the the current `input` and output a tuple of the `next_state` and the current `output` . 


````python
from abc import ABC, abstractmethod

class StateMachine(ABC):
    # StateMachine is an abstract class
    # cannot be instantiated, only inherited
    # CANNOT DO: sm = StateMachine()
    # we need to inherit, eg: CokeMachine(StateMachine), then CokeMachine implements start_state + get_next_values, then we can instantiate cm = CokeMachine()
    def __init__(self):
        self.state = None
        
    def start(self) -> None:
        # this is like powering on the machine e.g: plug the vending machine to the wall socket and see its start state
        # this method set the `state` property using the value in `start_state
        self.state = self.start_state # this will call the start_state property in the child class
        
    def step(self, inp: Any) -> Any:
        # this method takes in the current input and returns the current output. 
        # This method should move the state machine to the next state based on the 
        # current input and its current state. You should call `get_next_values(state, inp)` in your implementation.

        # STEP 1: Figure out the logic --> given current state and inp, what is the possible next state and output?
        # this logic is implemented in get_next_values function in the subclass
        next_state, output = self.get_next_values(self.state, inp)

        # STEP 2: Actually update the machine's state
        self.state = next_state 
        return output
        
        
    def transduce(self, inp_list: Sequence[Any]) -> list[Any]:
        # this method calls `start()` to initialize the `state` with the `start_state` 
        self.start()
        output_list = []
        # and run the state machine by calling `step(inp)` for every item in the `inp_list`. 
        for inp in inp_list:
            # The method runs the state machine and produces the output list according to the number of input in the `inp_list` 
            output_list.append(self.step(inp))
            # or when the state machine terminates according to the output of `is_done()` method. 
            # This method should call `is_done()` to see if it should terminate at a particular state.
            if (self.is_done()):
                break
        return output_list
        

    @property
    @abstractmethod # whoever inherit this class, SHOULD implement start_state property logic
    def start_state(self) -> Any:
        pass 

    # this is the transition logic
    @abstractmethod # whoever inherit this class, SHOULD implement get_next_values
    def get_next_values(self, state: Any, inp: Any) -> tuple[Any, Any]:
        pass

    def done(self, state: Any) -> bool:
        return False

    def is_done(self) -> bool:
        return self.done(self.state)


# sm = StateMachine()

# class A(StateMachine):
#     def a(self):
#         print("A")

# a = A()
````

````python
class CokeMachine(StateMachine):
    # need to implement this as stated in StateMachine
    @property
    def start_state(self) -> int:
        # this is a computed property
        # so just return a value, do not set any internal attr _start_state or whatever
        return 0

    # need to implement this as stated in StateMachine
    def get_next_values(self, state: int, inp: int) -> tuple[int, tuple[int, str, int]]:
        # this method just compute the logic transition only
        # it DOES NOT update self.state, it DOES NOT update the machine
        # return format: a tuple of next_state, output
        # output format: (money_in_machine, dispensed_drink, change)

        if (state == 0 and inp == 50):
            return (1, (50, "--", 0)) # 50c inside machine, no drink dispensed, 0 change given
        if (state == 0 and inp == 100):
            return (0, (0, "coke", 0)) 
        if (state == 1 and inp == 50):
            return (0, (0, "coke", 0))
        if (state == 1 and inp == 100):
            return (0, (0, "coke", 50))

        # if we reach here, that means we are receiving illegal inp, e.g: 10c coin, etc (not 50/100)
        # so we just return it as change and remain in state
        # what if we are at state 1, --> 50c inside, then we give 10c
            # choose to remain instate 1, keep 50c inside, but give 10c change
        return (state, (50 if state == 1 else 0, "--", inp)) 
````

In the code template below, we have provided a general getter and setter implementation for the computed property `start_state`. In this task, you need to set this computed property in the initializer `__init__()`.

````python
class SimpleAccount(StateMachine):
    @property
    def start_state(self) -> Number:
        return self._start_state
    
    @start_state.setter
    def start_state(self, value: Number) -> None:
        self._start_state: Number = value

    def __init__(self, balance: Number) -> None:
        super().__init__()
        self.start_state = balance # use the setter to set the start_state

    def get_next_values(self, state: Number, inp: Number) -> tuple[Number, Number]:
        # DO NOT use any self.state or self.anything in this method
        # for your computation, simply map out the logic using the args only: which is state and inp (NOT self.state or self.start_state or wtv)
        
        # The output of the state machine is the current balance after the transaction.
        # state is the money in the bank
        # inp can be + or - indicating deposit or withdrawal

        if inp >= 0:
            # this is a deposit
            next_state = state + inp # add the money
            output = state + inp # the money in the acc to print out
            return next_state, output

        if inp < 0:
            # this is a withdrawal
            if state >= 100: # money > $100, no withdrawal fee
                next_state = state + inp
                if next_state >= 0:
                    output = state + inp
                    return next_state, output
                # else:
                #     # insufficient money
                #     output = state # old balance
                #     return state, output

            # any withdrawal when the balance is less than \\$100  incurs a \\$5 charge
            # remember this method only encodes the SM logic and DOES NOT change the state yet
            if state < 100:
                next_state = state + inp - 5
                # check if there's enough balance
                if next_state >= 0:
                    output = state + inp - 5
                    return next_state, output
                # else: 
                #     # not enough money to withdraw + $5 cost
                #     output = state # old balance
                #     return state, output # return the same money back, unchanged

        return state, state # default return value, state unchanged, and return output as the current state     
        
````

**CS4.** Modify the `SimpleAccount` class in such a way that the account is automatically closed when the balance is negative. Make use of the `done()` method to define the terminating state. 

````python
class SimpleAccount(StateMachine):
    @property
    def start_state(self) -> Number:
        return self._start_state
    
    @start_state.setter
    def start_state(self, value: Number) -> None:
        self._start_state: Number = value

    def __init__(self, balance: Number) -> None:
       self.start_state = balance 

    def get_next_values(self, state: Number, inp: Number) -> tuple[Number, Number]:
        # this logic allows state to be -ve if we withdraw too much
        if inp < 0 and state < 100:
            # withdrawal with penalty
            next_state = state + inp - 5
            output = next_state
            return next_state, output
        next_state = state + inp
        output = next_state
        return next_state, output

    def done(self, state: Number) -> bool:
        return state < 0 
````

The class `MapSM` should implement `StateSpaceSearch` class which is an abstract class inheriting from `StateMachine` abstract class. The `StateSpaceSearch` abstract class requires `MapSM` to implement the properties:

````python
from abc import abstractmethod

class StateSpaceSearch(StateMachine):
    
    @property
    @abstractmethod
    def statemap(self) -> dict:
        pass

    @property
    @abstractmethod
    def legal_inputs(self) -> set:
        pass

    @property
    def start_state(self) -> Any:
        return self.__start_state
    
    @start_state.setter
    def start_state(self, value: Any) -> None:
        self.__start_state = value
    
    def get_next_values(self, state: Any, inp: Any) -> tuple[Any, Any]:
        # `get_next_values(state, inp)`: this method produces the next state based on the input and the current state. 
        # The `inp` argument is the index of the next state to be visited. 
        # The next states are the neighbours of the current state based on the `inp`. 
        
        # eg: get_next_values("D", 2)
        # neighbours --> "D": ["A", "B", "F", "H"]
        # next_state = neighbours[2] ---> "F"
        # output = next_state ---> "F"
        # return "F", "F"

        # what happens if inp is 2 and state is "S"?
        # neighbours --> "S": ["A", "B"]
        # next_state = neighbours[2] --> index out of range error
        # need to handle this
        
        # STEP 1: find neighbours first
        neighbours = self.statemap.get(state, None) # get it from the statemap, which is an property of the child class
        if neighbours is not None:
            # STEP 2: if we have a valid neighbour to go to
            if inp in self.legal_inputs and inp < len(neighbours): # can we make this move?
                # STEP 3: if we can make this move:
                next_state = neighbours[inp]
                output = next_state
                return next_state, output
            
        # The state machine should remain in the current state if the input is not valid or when the current state has no neighbours.
        return self.get_next_values_invalid(state, inp)

        

    def get_next_values_invalid(self, state: Any, inp: Any) -> tuple[Any, Any]:
        return state, 'invalid input' 
````

**CS6.** Create a class to represent a node in a state-space search called `SearchNode`. The object instance of `SearchNode` has the following properties which should be initialized during instantiation.
- `state`: which is the state or the label of the node.
- `action`: which is the action that was taken to arrive at the node.
- `parent`: which is the parent search node from which the current search node can be reached. If a node has no parent, this property should be initialized to `None`.

The class has the following method:
- `path()`: which returns a list of `Step` from the root node to the current node. The `Step` object is a class and is defined in the template. A `Step` object has two properties: `action` and `state`. You should use recursion for this method `path()`. The base case is when the node has no parent. In this case the solution contains only one step which is the action and the state of the current node. The recursive case is when the current has a `parent` object. In this case, you have to traverse to the ancestors' node until it reach a node which has no `parent` object. 
- `in_path(state)`: which returns `True` if the state in the argument is in the path. *Hint: use recursion and the parent's `in_path()` method.*

````python
class Step:
    def __init__(self, action: Optional[int], state: str) -> None:
        self.action: Optional[int] = action
        self.state: str = state
    
    def __eq__(self, other) -> bool:
        return self.action == other.action and self.state == other.state
  
    def __str__(self) -> str:
        return f"action: {self.action:}, state: {self.state:}"
````

````python
class SearchNode:
    def __init__(self, action: Optional[int], state: str, parent: Optional[SearchNode]) -> None:
        self.state: str = state
        self.action: Optional[int] = action
        self.parent: Optional[SearchNode] = parent
  
    def path(self) -> list[Step]:
        # which returns a list of `Step` from the root node to the current node. The `Step` object is a class and is defined in the template. 
        # A `Step` object has two properties: `action` and `state`. 
        # You should use recursion for this method `path()`. 
        
        # The base case is when the node has no parent. 
        if self.parent is None:
            # create a step object in a list
            # with attr: action and state
            # In this case the solution contains only one step which is the action and the state of the current node. 
            return [Step(self.action, self.state)]
        
        # The recursive case is when the current has a `parent` object. In this case, you have to traverse to the ancestors' node until it reach a node which has no `parent` object. 
        return self.parent.path() + [Step(self.action, self.state)]
        # example: 
        # sn1 = SearchNode(1,"S", sn2)
        # sn2 = SearchNode(2, "A", sn3)
        # sn3 = SearchNode(None, "B", None)
        # B --> A --> S 
        # sn1.path()
            # self.parent is "A" (sn2)
            # we go recursive case, and compute sn2.path()
            # sn2.path()
                # self.parent is "B" (sn3)
                # we go recursive case again, and compute sn3.path()
                # sn3.path()
                    # self.parent is NONE
                    # we go to the base case that returns
                    # [Step(None, "B")]
                    # this return to sn2.path() function
            # in sn2.path function, we do: [Step(None,"B")] + [Step(sn2.action: 2,sn2.state: "A")] --> [Step("None", B), Step(2, "A")]
        # in sn1.path function, we do: [Step("None", B), Step(2, "A")] + [Step(sn1.action: 1, sn1.state: "S")] ---> [Step("None", B), Step(2, "A"), Step(1, "S")]
        
  
    def in_path(self, state: str) -> bool:
        if self.state == state:
            return True
        elif self.parent is None:
            return False
        else:
            return self.parent.in_path(state)
  
    def __eq__(self, other) -> bool:
        if self is None and other is None:
            return True
        elif self is None:
            return False
        elif other is None:
            return False
        else:
            return self.state == other.state and self.parent == other.parent and \
                   self.action == other.action
````
